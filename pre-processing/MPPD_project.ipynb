# Part 1

## 1- Load Data

1.1- Load **houses.csv** to a Pandas DataFrame

````python
houses_df = pd.read_csv("houses.csv")
````

1.2- Print dataset total number of `observations` and `variables`

````python
dimensions = houses_df.shape
print("Observations: " + str(dimensions[0]))
print("Variables: " + str(dimensions[1]))
````

---

### Please find below the subset of columns we are going to consider for the rest of the assignment

````python
columns_list = ['FullBath',
                'TotRmsAbvGrd',
                'Fireplaces',
                'GarageYrBlt',
                'GarageCars',
                'GarageArea',
                'LotFrontage',
                'WoodDeckSF',
                'OpenPorchSF',
                'SaleType',
                'SaleCondition',
                'SalePrice']
````

1.3- Create a new dataframe which is a subset of the origin dataframe based on the columns listed above.

````python
h_df = houses_df[columns_list]
h_df.columns
````

## 2- Creating Lables

2.1- Create the `target` column based on `SalePrice`. The split should be done using the median value to create 2 new buckets. `Min->Median` bucket should have assigned the value `0` while the other bucket (`Median->Max`) value should be `1`.

Note: you are free to decide the buckets boundaries

````python
sale_price_median = h_df["SalePrice"].median()
````

````python
def target_builder(row):
     return 1 if row["SalePrice"] >= sale_price_median else 0
````

````python
h_df["target"] = h_df.apply(lambda row: target_builder(row), axis=1)
````

````python
h_df.head()
````

## 3- Handling Missing Values

3.1- List the amount of missing values per column

````python
h_df.isna().sum()
````

3.2- Take care of the missing values in the column `LotFrontage`

````python
#As we are only considering the subset of columns, for the missing values we are using the mean of all the non null values of Neighborhood_dict
h_df['LotFrontage'].fillna(h_df['LotFrontage'].mean(), inplace=True)
h_df.isna().sum()
````

## 4- Handling Categorical Data

4.1- Split categorical feature into a `df_categorical` dataframe

````python
#categorical features: SaleCondition,SaleType
columns = ['SaleCondition', 'SaleType']
df_categorical = h_df[columns]
df_categorical
````

4.2- Apply OHE to `SaleType`

````python
df_categorical = pd.get_dummies(df_categorical, columns=['SaleType'])
    
    
````

````python
df_categorical
````

## 5- Feature Scaling

5.1- Apply feature scaling to the variable `GarageArea`. Make sure that the new range fall between `-1/3` and `3`.

````python
# Instanciate the min-max scaler object
mmscaler = MinMaxScaler(feature_range=(-1/3, 3), copy=True)
````

````python
h_df['GarageArea_scaled'] = mmscaler.fit_transform(h_df[['GarageArea']])
h_df.head()
````

## Preparing for baseline
It will be made the minimum changes as possible in order to create a baseline for posterior evaluation on model performance

### Removing NaN

````python
#Removing the nan from LotFrontage
#As we are only considering the subset of columns, for the missing values we are using the mean of all the non null values of Neighborhood_dict
df['LotFrontage'].fillna(df['LotFrontage'].mean(), inplace=True)
df.isna().sum()
````

````python
#Removing the nan from MasVnrType
#setting Nan as Unknown
df['MasVnrType'].fillna('Unknown',inplace=True)
df.isna().sum()
````

````python
#Removing the nan from BsmtFinType1
#setting Nan as Unknown
df['BsmtFinType1'].fillna('Unknown',inplace=True)
df.isna().sum()
````

### Converting features for baseline
As is going to be used a numerical method on our classifier, it must convert all object data types to numerical.

It will be performed the minimum of changes as possible in order tokeep the dataset as the same as it was given. 


````python
#setting to category
df['MasVnrType'] = df['MasVnrType'].astype('category')

#label Enconding
df['MasVnrType_cat'] =df['MasVnrType'].cat.codes
df.dtypes
````

````python
#setting to category
df['Neighborhood'] = df['Neighborhood'].astype('category')

#Label Encoding
df['Neighborhood_cat'] =df['MasVnrType'].cat.codes
df.dtypes
````

````python
#setting to category
df['BsmtFinType1'] = df['BsmtFinType1'].astype('category')

#Label Encoding
df['BsmtFinType1_cat'] =df['BsmtFinType1'].cat.codes
df.dtypes
````

## Baseline

````python
columns = ['YearBuilt','LotFrontage', 'MasVnrType_cat','Neighborhood_cat', 'BsmtFinType1_cat','BsmtFinSF1']
baseline = train_and_evaluation(df[columns], df.target)
print("Baseline: " + str(baseline))
````

## Transformation 1

The `YearBuilt` is a spcecial case, because, although it is numerical, it should be handled as a category.

Under this context the year a house was built does not implies an order

````python
columns = ['YearBuilt','LotFrontage', 'MasVnrType_cat','Neighborhood_cat', 'BsmtFinType1_cat','BsmtFinSF1']
#Applying One Hot Enconding
df_year = pd.get_dummies(df[columns], columns=['YearBuilt'])
````

````python
accuracy = train_and_evaluation(df_year, df.target)
print("Baseline: " + str(baseline))
print("accuracy: " +str( accuracy))
````

**New baseline**
